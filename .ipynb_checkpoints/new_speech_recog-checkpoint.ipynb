```
from data_loaders import data_loader
from trainer import train
from torchvision import models, datasets, transforms
import torch
```

```
PATH = 'datasets/speech_command_data/'
train_path = f'{PATH}train'
valid_path = f'{PATH}valid'
test_path = f'{PATH}test'
```

```
train_dl, valid_dl, test_dl = data_loader.get_data_loaders(train_path, valid_path, test_path, 32, 0)
```

```
tfms = train_transforms = transforms.Compose([
            transforms.Resize((224,224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.2582, 0.1298, 0.3936], std=[0.0569, 0.2010, 0.0886])
        ])
train_ds = datasets.ImageFolder(train_path, tfms)
valid_ds = datasets.ImageFolder(valid_path, tfms)

train_dl = torch.utils.data.DataLoader(dataset=train_ds, batch_size=32, shuffle=True, num_workers=0)
valid_dl = torch.utils.data.DataLoader(dataset=valid_ds, batch_size=32, shuffle=True, num_workers=0)
```

```
model = models.resnet34(pretrained=True)
```

```
for param in model.parameters():
  param.requires_grad=False
```

```
model.fc = torch.nn.Linear(512,10)
```

```
criterion = torch.nn.CrossEntropyLoss()
```

```
trainer = train.Model(model, train_dl, valid_dl, None, criterion)
```

```
next(trainer.model.parameters()).is_cuda
```

```
optimizer = torch.optim.Adam(model.parameters(), lr=1e-2)
```

```
for i in range(3):
  print(f'Doing Epoch {i}')
  train_loss = trainer.train_epoch(optimizer, 'cuda:0')
  print('Done training')
  val_loss, val_acc = trainer.valid_epoch('cuda:0')
  print(f'Epoch {i}: Train loss:{train_loss.item()} Valid loss:{val_loss.item()} Valid accuracy:{val_acc.item()}')
```

```
for param in trainer.model.parameters():
  param.requires_grad=True
```

```
#Loading checkpoint
saved_state = torch.load('saved_state')
```

```
trainer.model.load_state_dict(saved_state['state_dict'])
```

```
optimizer.load_state_dict(saved_state['optimizer'])
```

```
train_loss, train_loss_float, train_loss_float_item = trainer.train_epoch(optimizer, 'cuda:0')
print(f'Train loss:{train_loss} train_loss_float:{train_loss_float} float_item:{train_loss_float_item}')
```

```
for i in range(2):
  print(f'Doing Epoch {i}')
  train_loss_float, train_loss_float_item = trainer.train_epoch(optimizer, 'cuda:0')
  print('Done training')
  print(f'float_loss:{train_loss_float} float_loss_item:{train_loss_float_item}')
  val_loss, val_acc = trainer.valid_epoch('cuda:0')
  print(f'Epoch {i}: Valid loss:{val_loss} Valid acc:{val_acc}')
```

```
state = {
    'state_dict':trainer.model.state_dict(),
    'optimizer':optimizer.state_dict()
}
```

```
torch.save(state, 'saved_state')
```
