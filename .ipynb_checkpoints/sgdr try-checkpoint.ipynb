```python
import torch
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
opt_lrs = []
```

```python
n_cycles = 4
cycle_len = 1
cycle_mult = 2
iters = 235
```

```python
for curr_cycle in range(n_cycles):
    n_epochs = cycle_len * (cycle_mult ** curr_cycle)
    optimizer = torch.optim.SGD(params=[torch.rand((5,5), torch.rand(5,2))], lr=1e-2, momentum=0.9)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, n_epochs*iters)
    scheduler.step()
    for curr_epoch in range(n_epochs):
        for curr_iter in range(iters):
            opt_lrs.append(optimizer.state_dict()['param_groups'][0]['lr'])
            optimizer.step()
            scheduler.step()
```
