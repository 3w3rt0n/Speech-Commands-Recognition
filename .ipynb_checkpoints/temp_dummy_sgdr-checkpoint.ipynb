```python
import torch
from torchvision import models, transforms, datasets
from data_loaders import data_loader
from trainer import train
```

```python
tfms = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize([0.25816761091643686, 0.1297258264878217, 0.39352345186121324], [0.056850156597062655, 0.20094163184072458, 0.08856581145641851])
])
```

```python
train_ds = datasets.ImageFolder('datasets/speech_command_data/train/', transform=tfms)
valid_ds = datasets.ImageFolder('datasets/speech_command_data/valid/', transform=tfms)
test_ds = datasets.ImageFolder('datasets/speech_command_data/test/', transform=tfms)
```

```python
train_dl = torch.utils.data.DataLoader(train_ds, batch_size=64, shuffle=True)
valid_dl = torch.utils.data.DataLoader(valid_ds, batch_size=64, shuffle=True)
test_dl = torch.utils.data.DataLoader(test_ds, batch_size=64, shuffle=True)
```

```python
model = models.resnet34(pretrained=True)
```

```python
criterion = torch.nn.CrossEntropyLoss()
```

```python
trainer = train.ModelTrainer(model, train_dl, valid_dl, test_dl, criterion)
```
